#Criação do Dataset

```python
import pandas as pd
import numpy as np
import statistics
import math
import random
```

```python
# dados = [150, 150, 151, 152, 152, 153, 154, 155, 156, 157, 157, 157, 159, 159, 160, 161, 161, 162, 163, 164, 166, 166, 167, 168, 169, 170, 170, 171]
dados = [random.randint(160, 190) for _ in range(100)]
```

```python
intervalos = [(145, 150), (150, 155), (155, 160), (160, 165), (165, 170), (170, 175), (175, 180), (180, 185), (185, 190), (190, 195), (195, 200)]
```

```python
def encontrar_intervalo(valor):
    for intervalo in intervalos:
        if intervalo[0] <= valor < intervalo[1]:
            return intervalo
    return None
```

```python
intervalos_encontrados = [encontrar_intervalo(valor) for valor in dados]
```

```python
dataset = pd.DataFrame({'Valor': dados, 'Intervalo': intervalos_encontrados})
dataset
```

```python
new_dataset = dataset.groupby('Intervalo').size().reset_index(name='Quantidade')
new_dataset
```

```python
pontos_medios = []
```

```python
for i in new_dataset['Intervalo']:
  Li = i[1]
  li = i[0]

  pontos_medios.append((Li + li) / 2)

pontos_medios
```

```python
new_dataset['pontos_medios'] = pontos_medios
new_dataset
```

```python
new_dataset['fi.xi'] = new_dataset['Quantidade'] * new_dataset['pontos_medios']
new_dataset
```

```python
Fi = []

for q, quantidade in enumerate(new_dataset['Quantidade']):
  if len(Fi) == 0:
    Fi.append(quantidade)
  else:
    valor_anterior = Fi[q - 1 if len(Fi) > 0 else 0]
    Fi.append(quantidade + valor_anterior)

Fi
```

```python
new_dataset['Fi'] = Fi
new_dataset
```

#Media

```python
def media(dataset):
  t_fi_xi = dataset['fi.xi'].sum()
  t_quantidade = dataset['Quantidade'].sum()

  return t_fi_xi / t_quantidade
```

```python
media_aritmetica = media(new_dataset)
media_aritmetica
```

#Mediana

```python
mediana = statistics.median(dados)
mediana
```

```python
#maneira de identificar a qual intervalo a mediana pertence
for i, intervalo in enumerate(new_dataset['Intervalo']):
  if mediana > intervalo[1]:
    continue
  elif mediana < intervalo[1] and mediana >= intervalo[0]:
    intervalo_m = i

intervalo_m
```

```python
l = new_dataset['Intervalo'][intervalo_m][0]
soma_fi = new_dataset['Quantidade'].sum() / 2
h = new_dataset['Intervalo'][intervalo_m][1] - new_dataset['Intervalo'][intervalo_m][0]
fi = new_dataset['Quantidade'][intervalo_m]
f_ant = new_dataset['Fi'][intervalo_m - 1]
```

```python
md = l + (((soma_fi - f_ant) * h) / fi)
md
```
