##Exercicio 2

```python
#Criao do dataset
dataset_2 = {
    'ano_graduacao': ['1', '2', '3', '4', 'Total'],
    'matriculas_marco': [70, 50, 47, 23, 190],
    'matriculas_novembro': [65, 48, 40, 22, 175],
}
dataset_2
```

```python
df2 = pd.DataFrame(dataset_2)
df2
```

```python
df2['coeficiente_evasao'] = (df2['matriculas_marco'] - df2['matriculas_novembro']) / df2['matriculas_marco']
df2
```

```python
df2['taxa_evasao'] = df2['coeficiente_evasao'] * 100
df2
```
